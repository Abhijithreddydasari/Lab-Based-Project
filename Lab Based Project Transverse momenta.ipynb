## This code contains curve fitting for tranverse momenta for different nuclei on a 'C' target

# Tranverse Momenta 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from lmfit import Model
from math import log, sqrt
```

```python
data1 = pd.read_csv('plot-data_t1.csv')
x1 = np.array(data1.iloc[:,0])
y1 = np.array(data1.iloc[:,1])
```

```python
# Gaussian function
def gaussian(x, a, b, c):
    return a*np.exp(-(x-b)**2/(2*c**2))

# Lorentzian function
def lorentzian(x, x0, gamma, A):
    return (A*gamma**2)/((x-x0)**2+gamma**2)
```

```python
popt1, pcov1 = curve_fit(gaussian, x1, y1, p0 =[450,0,180])
A_opt1, mu_opt1, sig_opt1 = popt1
plt.scatter(x1, y1, label='Experimental data')
x_model = np.linspace(min(x1), max(x1), 1000)
plt.plot(x_model, gaussian(x_model, popt1[0], popt1[1], popt1[2]), 'r-', label='Gaussian fit')
plt.legend()
plt.xlabel('Transverse momentum $P_x$ ${(MeV/c)}$')
plt.ylabel('Counts measured')
plt.title('Gaussian Fitting for Transverse momentum')
plt.show()

residuals = y1 - gaussian(x1, popt1[0], popt1[1], popt1[2])
ss_res = np.sum(residuals**2)
print("Cost function value for Gaussian fitting: ", ss_res)
fwhm_g = abs(2*sqrt(2*log(2))*popt1[2])
print("FWHM for Gaussian is ", fwhm_g)
```

```python
popt1, pcov1 = curve_fit(lorentzian, x1, y1, p0 = [180,0,450])
A_opt1, mu_opt1, sig_opt1 = popt1
plt.scatter(x1, y1, label='Experimental data')
x_model = np.linspace(min(x1), max(x1), 1000)
plt.plot(x_model, lorentzian(x_model, popt1[0], popt1[1], popt1[2]), 'r-', label='Lorentzian fit')
plt.legend()
plt.xlabel('Transverse momentum $P_x$ ${(MeV/c)}$')
plt.ylabel('Counts measured')
plt.title('Lorentzian Fitting for Transverse momentum')
plt.show()

residuals = y1 - lorentzian(x1, popt1[0], popt1[1], popt1[2])
ss_res = np.sum(residuals**2)
print("Cost function value for Lorentzian fitting: ", ss_res)
fwhm_l = abs(2*popt1[1])
print("FWHM for Lorentzian is ", fwhm_l)
```
